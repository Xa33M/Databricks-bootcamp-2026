```python
from bronze_layer_config import INGESTION_CONFIG
```

```python
%sql
USE CATALOG workspace;
USE SCHEMA bronze_layer;
```

```python
for item in INGESTION_CONFIG:

    print(f'ingesting from {item['source']} source system --> bronze_layer.{item['table']}')

    df = (
            spark.read
                .option('header', 'true')
                .option('inferSchema', 'true')
                .csv(item['path'])
    )

    (
        df.write
            .mode('overwrite')
            .format('delta')
            .saveAsTable(f'bronze_layer.{item['table']}')
    )
```
